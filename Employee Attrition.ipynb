**DISTRIBUTION OF EMPLOYEE ATTRITION IN THE COMPANY

```python
labels = 'Attrition NO','Attrition YES'
data['Attrition'].astype(str).value_counts().plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,                                
                            )

plt.title('Distribution of Employee Attrition in the Company ', y=1.12) 
plt.axis('equal') 
# add legend
plt.legend(labels=labels, loc='upper left') 
 # show plot
plt.show()                     
```

**Analysis of Business Travel Feature

```python
var = data.groupby("BusinessTravel")['Attrition'].value_counts(normalize=False).unstack()

var.plot(kind='bar',alpha=1, stacked='True')

plt.title('Business Travel VS Attrition')
plt.ylabel('Number of Employee')
plt.show()
```

**From the above data it is clear that Employees who travel rarely have more attrition rate followed by Employees who travel frequently
    
  

**Analysis of Work Experience

1.YearsAtCompany
2.YearsInCurrentRole
3.YearsSinceLastPromotion
4.YearsWithCurrManager
5.TotalWorkingYears

```python
var1 = data.groupby("YearsAtCompany")['Attrition'].value_counts(normalize=False).unstack()

var1.plot(kind='bar', stacked='False',figsize=(10,6))

plt.title('Years At Company of Employee')
plt.ylabel('Number of Employees')
plt.show()
```

**It is observed that the newly joined employees quit their jobs most.

```python
var3 = data.groupby("YearsInCurrentRole")['Attrition'].value_counts(normalize=False).unstack()
var4 = data.groupby("YearsSinceLastPromotion")['Attrition'].value_counts(normalize=False).unstack()


fig = plt.figure() # create figure

ax0 = fig.add_subplot(121) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(122) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
var3.plot(kind='bar', stacked='False',figsize=(20,6), ax=ax0) # add to subplot 1
ax0.set_title('Same Role')
ax0.set_xlabel('Years In Current Role')
ax0.set_ylabel('Number of Employees')

# Subplot 2: Line plot
var4.plot(kind='bar', stacked='False',figsize=(20,6), ax=ax1) # add to subplot 2
ax1.set_title ('Last Promotion')
ax1.set_ylabel('Number of Employees')
ax1.set_xlabel('Years Since Last Promotion')

plt.show()
```

**From the above two plots, it is very clear that Employees who are in same postion or not getting promoted tend to leave the company most.

```python
var5 = data.groupby("YearsWithCurrManager")['Attrition'].value_counts(normalize=False).unstack()
var6 = data.groupby("TotalWorkingYears")['Attrition'].value_counts(normalize=False).unstack()


fig = plt.figure() # create figure

ax0 = fig.add_subplot(121) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(122) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
var5.plot(kind='bar', stacked='False',figsize=(20,6), ax=ax0) # add to subplot 1
ax0.set_title('YearsWithCurrManager')
ax0.set_xlabel('YearsWithCurrManager')
ax0.set_ylabel('Number of Employees')

# Subplot 2: Line plot
var6.plot(kind='bar', stacked='False',figsize=(20,6), ax=ax1) # add to subplot 2
ax1.set_title ('TotalWorkingYears')
ax1.set_ylabel('Number of Employees')
ax1.set_xlabel('TotalWorkingYears')

plt.show()
```

**From the above two plots, it is clear that the new joiness have left the job early than experienced ones.
