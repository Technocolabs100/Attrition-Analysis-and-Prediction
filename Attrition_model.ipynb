```python
df["Attrition"] = df["Attrition"].replace({"Yes":1, "No":0})
df["BusinessTravel"] = df["BusinessTravel"].replace({'Travel_Rarely':0, 'Travel_Frequently':1, 'Non-Travel':2})
df["Department"] = df["Department"].replace({'Sales':0, 'Research & Development':1, 'Human Resources':2})
df["EducationField"] = df["EducationField"].replace({'Life Sciences':0, 'Other':1, 'Medical':2, 'Marketing':3,
       'Technical Degree':4, 'Human Resources':5})
df["Gender"] = df["Gender"].replace({'Female':0, 'Male':1})
df["JobRole"] = df["JobRole"].replace({'Sales Executive':0, 'Research Scientist':1, 'Laboratory Technician':2,
       'Manufacturing Director':3, 'Healthcare Representative':4, 'Manager':5,
       'Sales Representative':6, 'Research Director':7, 'Human Resources':8})
df["MaritalStatus"] = df["MaritalStatus"].replace({'Single':0, 'Married':1, 'Divorced':2})
df["OverTime"] = df["OverTime"].replace({'Yes':1, 'No':0})
```

```python
df[non_number_index].head()
```

```python
df = df.drop(columns="Over18")
```

### Features processing

```python
features = df.drop(columns=["Attrition"])
target = df["Attrition"]
```

### Normalisation & PCA

```python
# Imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
normalized_X = scaler.fit_transform(features)
```

### How many components to chose

```python

pca = PCA()
X_pca = pca.fit_transform(normalized_X)

# explained Variance
explained_variance_ratio = pca.explained_variance_ratio_

# Cumulative variance
cumulative_variance = np.cumsum(explained_variance_ratio)

# Graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Number of PC')
plt.ylabel('Cumulative variance')
plt.title('Number of Principal Components -> Cumulative Variance')
plt.grid(True)
plt.show()
```

```python
for i, v in enumerate(cumulative_variance):
    print(f"PC {i+1}: Cumulative Variance = {v:.2f}")
```

```python
# PC 23 -> 95% Cum. Variance
pca = PCA(n_components=23)
X_pca = pca.fit_transform(normalized_X)
len(X_pca[0])
```

```python
X = pd.DataFrame(X_pca)
X.head()
```

```python
corr_matrix = X.corr()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
df_pca = pd.concat([X,target], axis=1)
df_corr = df_pca.corr()
df_pca.head()
```

```python
plt.figure(figsize=(35, 35))
sns.heatmap(df_corr, annot=True)
plt.show()
```

### Models

```python
y = target.values.reshape(-1, 1) # Target
print(y[:10])
```

### Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=20)
```

```python
print(f"Feature Train shape : {len(X_train)} - Feature Test shape : {len(X_test)}")
```

## Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

### I-Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
# Metrics import
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
# Accuracy of the model
print(accuracy_score(y_true=y_test, y_pred=y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_true=y_test, y_pred=y_pred)
print(confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

### II- Decision Tree
