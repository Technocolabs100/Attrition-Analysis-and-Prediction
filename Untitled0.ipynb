```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Loading the dataset
file_path = '/content/HR-Employee-Attrition.csv'
data = pd.read_csv(file_path)
```

```python
data=data.dropna()
```

```python
data.duplicated().sum()
```

```python
categorical_columns = data.select_dtypes(include='object').columns
print("\nCategorical columns to encode:", categorical_columns)
```

```python
label_enc = LabelEncoder()
onehot_enc = OneHotEncoder(sparse=False)

for column in categorical_columns:
    if data[column].nunique() == 2:  # Binary categorical columns
        data[column] = label_enc.fit_transform(data[column])
    else:  # Multi-class categorical columns
        onehot_encoded = onehot_enc.fit_transform(data[[column]])
        onehot_data = pd.DataFrame(onehot_encoded, columns=[f"{column}_{cat}" for cat in onehot_enc.categories_[0]])
        data = pd.concat([data, onehot_data], axis=1).drop(column, axis=1)

print("\nData after encoding:")
print(data.head())
```

```python
new_file_path = '/content/HR-Employee-Attrition-preprocessed.csv'
data.to_csv(new_file_path, index=False)
print(f"\nPreprocessed dataset saved to {new_file_path}")
```

```python
def plot_attrition_rates(column):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, hue='Attrition', data=data)
    plt.title(f'Attrition vs {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='Attrition', labels=['No', 'Yes'])
    plt.show()

# Job Satisfaction
plot_attrition_rates('JobSatisfaction')

# Work Life Balance
plot_attrition_rates('WorkLifeBalance')

# Relationship Satisfaction
plot_attrition_rates('RelationshipSatisfaction')
```

```python
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
target_column = 'Attrition'
X = data.drop(columns=[target_column])
y = data[target_column]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```
